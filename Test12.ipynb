```python
import torch
import torch.nn as nn
import torch.optim as optim

# Предположим, у вас есть данные X_train и метки y_train

# Преобразование в тензоры PyTorch
X_train_tensor = torch.Tensor(X_train_normilaze)
y_train_tensor = torch.Tensor(y_train).view(-1, 1)

# Определение архитектуры нейронной сети
class YourNeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(YourNeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, 32)
        self.fc3 = nn.Linear(32, output_size)

        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.sigmoid(x)
        return x

# Инициализация модели
input_size = X_train_normilaze.shape[1]
hidden_size = 64
output_size = 1
model = YourNeuralNetwork(input_size, hidden_size, output_size)

# Определение функции потерь и оптимизатора
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Обучение модели
num_epochs = 50
batch_size =128
for epoch in range(num_epochs):
    for i in range(0, len(X_train), batch_size):
        inputs = X_train_tensor[i:i + batch_size]
        targets = y_train_tensor[i:i + batch_size]
        # Forward pass
        outputs = model(inputs)

        # Рассчет функции потерь
        loss = criterion(outputs, targets)
        #loss = criterion(outputs.squeeze(), y_train_tensor)
        # Обратный проход и оптимизация
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        

        # Вывод прогресса
        
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}')
```
